## Function

```python
import json
def load(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        data = json.load(file)
    return data

def dump(filename,data):
    with open(filename, 'w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)

def getTermSet(dic):
    tmp = set()
    for i in dic:
        tmp.add(i['term'])
    return tmp
```
